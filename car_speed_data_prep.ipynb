```
TRAIN_VIDEO = ('train.mp4')
TEST_VIDEO = ('test.mp4')
CLEAN_IMGS_TRAIN = ('/content/gdrive/My Drive/ColabNotebook/comma/train_imgs/')
CLEAN_IMGS_TEST = ('/content/gdrive/My Drive/ColabNotebook/comma/test_imgs/')

train_frames = 20400
test_frames = 1
```

```
train_y = list(pd.read_csv('/content/gdrive/My Drive/ColabNotebook/comma/train.txt', header=None, squeeze=True))
assert(len(train_y)==train_frames)
```

```
def dataset_constructor(video_loc, img_folder, tot_frames, dataset_type):
    meta_dict = {}

    tqdm.write('reading in video file...')
    tot_frames = train_frames
    cap =  cv2.VideoCapture(video_loc)
    idx = 0
    while(cap.isOpened()):
        if idx == train_frames:
            break
        # Capture frame-by-frame
        ret, frame = cap.read()  
        if ret:           
            img_path = os.path.join(img_folder, str(idx)+'.jpg')
            frame_speed = float('NaN') if dataset_type == 'test' else train_y[idx]
            print(img_path, idx, frame_speed)
            meta_dict[idx] = [img_path, idx, frame_speed]
            idx+=1
            cv2.imwrite(img_path, frame)

            

    # When everything done, release the capture
    cap.release()

   
    return meta_dict
```

```
meta_df = dataset_constructor(TRAIN_VIDEO, CLEAN_IMGS_TRAIN, train_frames, 'train')
meta_data = pd.DataFrame.from_dict(meta_df, orient='index', columns = ['image_path', 'image_index', 'speed'])
meta_data.to_csv('train_meta.csv', index=False)
```
